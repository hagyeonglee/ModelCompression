```python
from util import get_clamped_psnr
def evaluate_siren(model,img_path):
    #dtype = torch.float32
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    results = {'bpp':[],'psnr_values': [], 'ssim_values': []}

    img = imageio.imread(img_path)
    img = transforms.ToTensor()(img).float().to(device)

    model.eval()
    with torch.no_grad():
        coordinates, features = util.to_coordinates_and_features(img)
        coordinates, features = coordinates.to(device), features.to(device)

    predicted = model(coordinates)
    psnr = get_clamped_psnr(predicted, features)

    results['psnr_values'].append(psnr)

    with torch.no_grad():
        img_recon = model(coordinates).reshape(img.shape[1], img.shape[2], 3).permute(2, 0, 1)
        save_image(torch.clamp(img_recon, 0, 1).to('cpu'), '/content/drive/MyDrive/22WinterIntern/' + f'/quant_reconstruction_ada.png')

    return results['psnr_values']
```

```python
from functools import partial
from aimet_common.defs import QuantScheme
from aimet_torch.adaround.adaround_weight import Adaround, AdaroundParameters
from aimet_torch.utils import create_fake_data_loader
from aimet_torch.cross_layer_equalization import equalize_model
def siren_quanteval_adaround(model_dir):
    model = Siren(
    dim_in=2,
    dim_hidden=28,
    dim_out=3,
    num_layers=10,
    final_activation=torch.nn.Identity(),
    w0_initial=30.0,
    w0=30.0
    ).to(torch.device('cuda'))

    state_dict = torch.load(model_dir)
    state = model.state_dict()
    state.update(state_dict)
    model.load_state_dict(state)

    input_shape = (10,2)

    equalize_model(model, input_shape)

    # create Quantization Simulation model with appropriate quant_scheme, supported schemes :
    quant_scheme = QuantScheme.post_training_tf_enhanced

    data_loader = create_fake_data_loader(dataset_size=64, batch_size=4, image_size=(10,2))

    # configuration parameters
    params = AdaroundParameters(data_loader, num_batches=4,default_num_iterations=50,default_reg_param=0.01,default_beta_range=(20, 2))

    adarounded_model = Adaround.apply_adaround(model,torch.rand(input_shape).to('cuda'),
                                               params,path="/content/drive/MyDrive/22WinterIntern/",
                                               filename_prefix='siren_ada',
                                               default_param_bw=8,
                                               default_quant_scheme=quant_scheme,
                                               default_config_file=None)

    sim_siren = QuantizationSimModel(adarounded_model,torch.rand(input_shape).to('cuda'),quant_scheme=quant_scheme,default_output_bw=8,default_param_bw=8)

    sim_siren.set_and_freeze_param_encodings(encoding_path="/content/drive/MyDrive/22WinterIntern/siren_ada.encodings")

    evaluate = partial(evaluate_siren)

    sim_siren.compute_encodings(evaluate, "/content/drive/MyDrive/22WinterIntern/coin/kodak-dataset/kodim03.png")

    psnr_vals = evaluate_siren(sim_siren.model,"/content/drive/MyDrive/22WinterIntern/coin/kodak-dataset/kodim03.png")
    psnr_val = util.mean(psnr_vals)

    print(f'Mean PSNR on quantized model is: [{psnr_val}]')

    print(sim_siren.model.net[9].activation.output_quantizer.bitwidth)
```

### [3] Quantization(INT8) -> Quantization(AdaRound) -> Quantization Aware Training

```python
#torch.autograd.set_grad_enabled(True)
```

```python
model_q = Siren(
    dim_in=2,
    dim_hidden=28,
    dim_out=3,
    num_layers=10,
    final_activation=torch.nn.Identity(),
    w0_initial=30.0,
    w0=30.0
    ).to(torch.device('cuda'))
```

```python
input_shape = (10,2)

equalize_model(model_q, input_shape)

# create Quantization Simulation model with appropriate quant_scheme, supported schemes :
quant_scheme = QuantScheme.post_training_tf_enhanced

data_loader = create_fake_data_loader(dataset_size=64, batch_size=4, image_size=(10,2))

# configuration parameters
params = AdaroundParameters(data_loader, num_batches=4,default_num_iterations=50,default_reg_param=0.01,default_beta_range=(20, 2))

adarounded_model = Adaround.apply_adaround(model_q,torch.rand(input_shape).to('cuda'),
                                            params,path="/content/drive/MyDrive/22WinterIntern/",
                                            filename_prefix='siren_ada_q',
                                            default_param_bw=8,
                                            default_quant_scheme=quant_scheme,
                                            default_config_file=None)

sim_siren_q = QuantizationSimModel(adarounded_model,torch.rand(input_shape).to('cuda'),quant_scheme=quant_scheme,default_output_bw=8,default_param_bw=8)

sim_siren_q.set_and_freeze_param_encodings(encoding_path="/content/drive/MyDrive/22WinterIntern/siren_ada_q.encodings")

evaluate = partial(evaluate_siren)

sim_siren_q.compute_encodings(evaluate, "/content/drive/MyDrive/22WinterIntern/coin/kodak-dataset/kodim03.png")
```

```python
trainer_sim_q = Trainer(sim_siren_q.model, lr=2e-4)
```

```python
img = imageio.imread("/content/drive/MyDrive/22WinterIntern/coin/kodak-dataset/kodim03.png")
img = transforms.ToTensor()(img).float().to('cuda')
```

```python
coordinates, features = util.to_coordinates_and_features(img)
```

```python
coordinates, features = coordinates.to('cuda'), features.to('cuda')
```

```python
trainer_sim_q.train(coordinates, features, num_iters=5000, show=True)
print(f'Best training psnr: {trainer_sim_q.best_vals["psnr"]:.2f}')
```
